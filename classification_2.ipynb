```python
#all imports
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from IPython.display import Image
%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data[:, [2,3]]
y = iris.target
label_list = np.unique(y)
print(label_list)
#setosa, versicolor and virginica
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
print("Label count overall: ", np.bincount(y))
print("Label count (train): ", np.bincount(y_train))
print("Label count (test): ", np.bincount(y_test))
#stratify = y ensures equal proportion of labels across train and test sets
```

```python
#standardize the features
ss = StandardScaler()
ss.fit(X_train) #estimate mean and std from training data
X_train_std = ss.transform(X_train)
X_test_std = ss.transform(X_test)
```

```python
model = Perceptron(eta0=0.01, random_state=1)
model.fit(X_train_std, y_train)
```

```python
#predictions
y_pred = model.predict(X_test_std)
print('Misclassified examples: {}'.format((y_test!=y_pred).sum()))
error = ((y_test!=y_pred).sum())/float(y_test.shape[0])
print('Error: {:.4f}%'.format(100*error))
print('Accuracy: {:.4f}%'.format(100*(1-error)))
```

```python
#computing accuracy using sklearn
print('Accuracy: {:.4f}% (sklearn)'.format(100*accuracy_score(y_test, y_pred)))
print('Accuracy: {:.4f}% (sklearn)'.format(100*model.score(X_test_std, y_test)))
```

```python
def plot_decision_boundary(X, y, classifier, test_idx=None,resolution=0.02):
    #basic setup
    markers = ('s','x','o','^','v' )
    colors = ('red','royalblue','springgreen','gray','lightcoral')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    #plot
    x1_min, x1_max = X[:,0].min() - 1, X[:,0].max() + 1
    x2_min, x2_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    z = z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, col in enumerate(np.unique(y)):
        plt.scatter(x = X[y==col, 0], y=X[y==col,1],
                   alpha=0.8, c=colors[idx],
                   marker = markers[idx], label=col,
                   edgecolor='black')
        
    #highlight test examples
    if test_idx:
        #plot all
        X_test, y_test = X[test_idx, :], y[test_idx]
        
        plt.scatter(X_test[:,0], X_test[:,1],
                   c='None', edgecolor='black', alpha=1.0,
                   linewidth=1, marker='o',
                   s=100, label='test_set')
```

```python
X_stacked_std = np.vstack((X_train_std, X_test_std))
y_stacked = np.hstack((y_train, y_test))
plot_decision_boundary(X = X_stacked_std,
                       y = y_stacked,
                       classifier = model,
                       test_idx = range(105,150))
plt.xlabel('petal length (standardized)')
plt.ylabel('petal width (standardized)')
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

```python
class LogisticRegressionModel(object):
    def __init__(self, lr=0.05, epochs=100, random_seed=42):
        self.lr = lr
        self.epochs = epochs
        self.random_seed = random_seed
    
    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_seed)
        self.weights = rgen.normal(loc = 0.0, scale=0.01,
                                  size = X.shape[1]+1)
        #extra weight for bias term
        self.cost_list = []
        
        for i in range(self.epochs):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = y - output #matrix of error for every training example
            self.weights[1:] += self.lr * X.T.dot(errors)
            self.weights[0] += self.lr * errors.sum()
            cost = (-y.dot(np.log(output)) - 
                   ((1-y).dot(np.log(1-output))))
            self.cost_list.append(cost)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.weights[1:]) + self.weights[0]
    
    def activation(self, z):
        return 1.0/(1.0 + np.exp(-np.clip(z,-200,200)))
    
    def predict(self, X):
        return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
```

```python
X_train_subset = X_train[(y_train==0) | (y_train==1)]
y_train_subset = y_train[(y_train==0) | (y_train==1)]

model2 = LogisticRegressionModel(lr=0.05, epochs=1000, random_seed=42)
model2.fit(X_train_subset, y_train_subset)
plot_decision_boundary(X=X_train_subset, y=y_train_subset,
                      classifier = model2)
plt.xlabel('petal length (standardized)')
plt.ylabel('petal width (standardized)')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Let's have a look at kernel approaches for linearly inseparable data! \
The simplest and most common example of linearly inseparable data is the XOR dataset, which we can conveniently generate when needed (as follows)

```python
np.random.seed(42)
X_xor = np.random.randn(200,2)
y_xor = np.logical_xor(X_xor[:,0] > 0, X_xor[:,1]>0)
y_xor = np.where(y_xor, 1, -1)
#100 examples will be assigned 1, others (also 100) -1
plt.scatter(X_xor[y_xor == 1, 0],
            X_xor[y_xor == 1, 1],
            c='blue',marker='x',
            label='1')
plt.scatter(X_xor[y_xor == -1, 0],
            X_xor[y_xor == -1, 1],
            c='red',marker='s',
            label='1')
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

A linear classifier clearly won't do justice to this dataset, but an SVM with a radial basis function (RBF) kernel might!

```python
svm = SVC(kernel='rbf', random_state=1, gamma=0.1, C=10.0)
svm.fit(X_xor, y_xor)
plot_decision_boundary(X_xor, y_xor, classifier=svm)
plt.title('Decision boundary for XOR data')
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

Time to go on to decision trees! \
The idea is to maximize information gain at each stage, which would be a function of some node-dependent impurity measure we choose to work with. Let's first quickly plot these impurity measures, and then come back to implementing a decision tree classifier.

```python
def entropy(p):
    val = -p*np.log2(p) - (1-p)*np.log2(1-p)
    return val
def class_error(p):
    val = 1 - np.max([p,1-p])
    return val
def gini(p):
    val = 2*p*(1-p)
    return val

interval = np.arange(0.0, 1.0, 0.01)
entropy = [entropy(p) if p!=0 else None for p in interval]
scaled_entropy = [e*0.5 if e else None for e in entropy]
class_error = [class_error(p) for p in interval]
fig = plt.figure()
ax = plt.subplot()

for i, label, ls, color, in zip([entropy, scaled_entropy, gini(interval), class_error],
                          ['Entropy', 'Entropy*0.5',
                           'Gini impurity',
                           'Misclassification error'],
                          ['-', '-', '--', '-'],
                          ['black', 'blue',
                           'red', 'green']):
                           line = ax.plot(interval, i, label=label,
                           linestyle=ls, lw=2, color=color)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=5, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='black', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='black', linestyle='--')
plt.ylim([0, 1.05])
plt.xlabel('p(i=1)')
plt.ylabel('impurity index')
plt.show()
```

```python
X_stacked = np.vstack((X_train, X_test))
y_stacked = np.hstack((y_train, y_test))

def tree_iris(max_depth, visualize=False):
    tree= DecisionTreeClassifier(criterion='gini',max_depth=max_depth,random_state=1)
    tree.fit(X_train, y_train)
    plot_decision_boundary(X_stacked, y_stacked, classifier=tree,
                           test_idx = range(105,150))
    plt.xlabel('petal length (in cm)')
    plt.ylabel('petal width (in cm)')
    plt.title('Decision Tree Classifier with max_depth={}'.format(max_depth))
    plt.tight_layout()
    plt.show()
    
    if visualize == True:
        plt.figure(figsize = (10,12))
        plot_tree(tree, filled=True,node_ids=True) #visualize decision tree
        plt.show()
    return

tree_iris(4, visualize=True)
```

What's the effect of changing max_depth in the above decision tree classifier? Have a look!

```python
for i in range(1,7):
    tree_iris(i)
```

```python
#Random Forest Classifier
def _forest(n_estimators):
    forest = RandomForestClassifier(criterion='gini',n_estimators=n_estimators,random_state=42, n_jobs=2)
    forest.fit(X_train,y_train)
    plot_decision_boundary(X_stacked, y_stacked, classifier=forest,
                           test_idx = range(105,150))
    plt.title('Random Forest Classifier with {} trees'.format(n_estimators))
    plt.xlabel('petal length (in cm)')
    plt.ylabel('petal width (in cm)')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()
    return

num_trees = [5,10,15,20,25,50,100]
for n_estimators in num_trees:
    _forest(n_estimators)
```

```python
#KNN classifier
distance_metric = ['Manhattan','Euclidean']
def _knn(n_neighbors, p):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, p=p, metric='minkowski')
    knn.fit(X_train, y_train)
    plot_decision_boundary(X_stacked,y_stacked,
                           classifier=knn, test_idx=range(105,150))
    plt.title('KNN Classifier with {} neighbors and {} distance'.format(n_neighbors, distance_metric[p-1]))
    plt.xlabel('petal length (standardized)')
    plt.ylabel('petal width (standardized)')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()
    return
```

```python
for num in range(1,11):
    _knn(num, 1) #manhattan
for num in range(1,11):
    _knn(num, 2) #euclidean
```
